```python
import pandas as pd
```

```python
df = pd.read_excel("JijiCarsRawDataFinal.xlsx")
head = df.head()
head
```

```python
# Describe the entire dataset
df.describe()
```

```python
# Removed the empty spaces
# df = df.dropna(inplace = True)
df.dropna(inplace = True)
df.describe()
```

```python
# Replace the unkown YOM witht he mode YOM
for car in df.index:
    if df.loc[car, "YOM"] == "Unkown":
        df.loc[car, "YOM"] = 2014
df.describe()
```

```python
# Remove unrealistic Year of Manufactururer from the data
for car in df.index:
    if df.loc[car, "YOM"] < 1800 and df.loc[car, "YOM"] > 2024:
        df.drop(car, inplace = True) 
df.describe()
```

```python
# Replacing the missing color entry with black
# as it turns out all the entries with numerical values of color, 
# the YOM value was the one copied on that entry.
for car in df.index:
    if df.loc[car, "Color"].isdigit():
        print(df.loc[car,["Model","YOM", "Color"]])
        df.loc[car, "Color"] = "Black"
```

```python
for car in df.index:
    if df.loc[car, "Transmission"].isdigit():
        print(df.loc[car,["Model","Mileage", "Transmission"]])
```

```python
# As it turns out all the values of Tramission that are numeric
# They were supposed to be in mileage section.
# Rectifying that 

for car in df.index:
    if df.loc[car, "Transmission"].isdigit():
        df.loc[car, "Mileage"] = df.loc[car, "Transmission"]
```

```python
# Repeating the 2nd above code, it should give us same value for bot entries.
for car in df.index:
    if df.loc[car, "Transmission"].isdigit():
        print(df.loc[car,["Model","Mileage", "Transmission"]])
```

```python
# Replace the Transmission value now with Auto
for car in df.index:
    if df.loc[car, "Transmission"].isdigit():
        df.loc[car, "Transmission"] = "Automatic"
```

```python
# Making sure that the Price column is numeric
a = 0

b = 0

for car in df.index:
    for c in [str(df.loc[car, "Price"]).split(" ")]:
        #print(c)
        if "KSh" in c:
            try:
                df.loc[car, "Price"] = c[1]
            except IndexError:
                df.loc[car, "Price"] = 1300000
                b +=1
        else:
            a +=1
            pass
print("Len of b", b)

print("Len of a", a) 

        
    
```

```python
# remove the comma in the price column
for car in df.index:
    df.loc[car, "Price"] = int(str(df.loc[car, "Price"]).replace(',', ''))
    
```

```python
# there was one entry with Used column as a digit. Replaced now
cl = 0
for car in df.index:
    if df.loc[car, "Used"].isdigit():
        df.loc[car,"Used"] = "Kenyan Used"
        cl+=1
        print(car)
        print(df.loc[car, ["Make","Used"]])
```

```python
# number of entries with miliage value provided
mi = 0
for car in df.index:
    if str(df.loc[car, "Mileage"]).isdigit():
        mi +=1
print(mi)
```

```python
# Removes all the duplicate data. 
# It does not mean that the data is wromg
# Just cars with exact same attributes
# Compare with the above dataframe description to see the difference
df.drop_duplicates(inplace = True)
df.describe()
```

```python
# Clean data
price_mean = df.loc[:, "Price"].mean()

print(price_mean)
```

```python
# Clean data with duplicates
price_mean = df.loc[:, "Price"].mean()
price_median = df.loc[:, "Price"].median()
price_mode = df.loc[:, "Price"].mode()
print("Mean:",price_mean)
print("Median:",price_median)
print("Mode",price_mode)
```
